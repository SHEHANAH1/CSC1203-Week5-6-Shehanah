```python
# Q1
numbers = [float(input(f"Enter number {i+1}: ")) for i in range(5)]
total = sum(numbers)
average = total / len(numbers)

maximum = max(numbers)
minimum = min(numbers)

print("Numbers:", numbers)
print("Sum:", total)
print("Average:", average)
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
# Q2

words = tuple(input(f"Enter word {i+1}: ") for i in range(5))

print("First word:", words[0])
print("Last word:", words[-1])
print("Total words:", len(words))

searchW = input("Enter a word to check:")
if searchW in words:
    print(f"'{searchW}' is in the tuple.")
else:
    print(f"'{searchW}' is not in the tuple.")
```

```python
# Q3
students = {}
for i in range(3):
    name = input(f"Enter student {i+1} name: ")
    score = float(input(f"Enter {name}'s score: "))
    students[name] = score

print(" students and their scores:")
for name, score in students.items():
    print(f"{name}: {score}")

max_student = max(students, key=students.get)
print(f"\nStudent with the highest score: {max_student} ({students[max_student]})")
```

```python
# Q4
list1 = [int(x) for x in input("Enter numbers for the list 1 separated by spaces: ").split()]
list2 = [int(x) for x in input("Enter numbers for the list 2 separated by spaces: ").split()]

set1 = set(list1)
set2 = set(list2)

union = set1 | set2
intersection = set1 & set2
difference = set1 - set2

print("Union:", union)
print("Intersection:", intersection)
print("Difference (set1 - set2):", difference)
```

```python
# Q5
names = [input(f"Enter name {i+1}: ") for i in range(5)]

names_tuple = tuple(names)
print("Tuple of names:", names_tuple)

names_dict = {name: len(name) for name in names}
print("Dictionary (name: length):", names_dict)

lengths_set = set(names_dict.values())
print("Set of unique name lengths:", lengths_set)
```
