```python
# Q1
import math

def calculate_area(radius):
    return math.pi * radius ** 2

circumference = lambda radius: 2 * math.pi * radius

radius = float(input("Enter the radius of a circle: "))

area = calculate_area(radius)
circ = circumference(radius)

print(f"Area: {area:.2f}")
print(f"Circumference: {circ:.2f}")
```

```python
# Q2
import re

def sum_numbers(numbers):
    return sum(numbers)

double = lambda total: total * 2

Input = input("Enter a string:")

found_numbers = re.findall(r'\d+', Input)

numbers = list(map(int, found_numbers))

total = sum_numbers(numbers)

doubledV = double(total)

print("Numbers found:", numbers)
print("Sum:", total)
print(f"Doubled value: {doubledV}")
```

```python
# Q3
import math

def square_numbers(numbers):
    return list(map(lambda n: n ** 2, numbers))

Input = input("Enter numbers separated by commas: ")

numbers_list = [float(num) for num in Input.split(",")]

squares = square_numbers(numbers_list)

max_square = max(squares)

sumSqares = sum(squares)

sqrt_sumSqares = math.sqrt(sumSqares)

print("Numbers:", numbers_list)
print("Squares:", squares)
print("Maximum square:", max_square)
print("Square root of sum of squares:", sqrt_sumSqares)
```

```python
# Q4
import re
import math

def average_score(scores):
    average = sum(scores) / len(scores)
    return (lambda x: x)(average)

password = input("Enter your password: ")

Up = re.search(r"[A-Z]", password)
Low = re.search(r"[a-z]", password)
Num = re.search(r"\d", password)

if Up and Low and Num:
    strength = "Strong"
else:
    strength = "Weak"

scores = input("Enter 5 scores separated by spaces: ")

scores_list = [float(score) for score in scores.split()]

average = average_score(scores_list)

rounded_average = math.ceil(average)

print("Password strength:", strength)
print("Rounded average score:", rounded_average)
```

```python
# Q5
import re
import math

def count_vowels(text):
    return sum(1 for char in text.lower() if char in "aeiou")

def word_lengths(words):
    return list(map(lambda w: len(w), words))

paragraph = input("Enter a paragraph:")

vowel_words = re.findall(r'\b[aeiouAEIOU]\w*', paragraph)

total_vowels = count_vowels(paragraph)

lengths = word_lengths(vowel_words)

average_length = round(sum(lengths) / len(lengths), 2) if lengths else 0

print("Words starting with vowels:", vowel_words)
print("Total vowels in paragraph:", total_vowels)
print("Average vowel-word length:", average_length)
```
